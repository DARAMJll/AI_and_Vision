```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
```

```python
def floodFillPostProcess(src, diff = (2,2,2)): #이웃에 있는 칼라 값과의 픽셀 차이가 diff만큼이면 같은 색(클러스터)으로 칠함, diff 키울수록 크게 모이고 줄일수록 원영상과 비슷해짐
    img = src.copy()
    rows, cols = img.shape[:2]
    mask = np.zeros(shape=(rows+2, cols+2), dtype = np.uint8)
    for y in range(rows):
        for x in range(cols):
            if mask[y+1, x+1] == 0: #0이면 그려라?
                r = np.random.randint(256)
                g = np.random.randint(256)
                b = np.random.randint(256)
                cv2.floodFill(img, mask, (x, y), (b, g, r), diff, diff) #-2, +2 사이 : 같은색으로 칠함 
    return img
```

```python
src = cv2.imread('./segmentation_rsc/flower.jpg')
srcrgb = cv2.cvtColor(src, cv2.COLOR_BGR2RGB)
dst = floodFillPostProcess(src, diff = (5, 5, 5))

plt.figure(figsize = (10, 5))

plt.subplot(1, 2, 1)
plt.axis('off')
plt.imshow(srcrgb)

plt.subplot(1, 2, 2)
plt.axis('off')
plt.imshow(dst)

plt.show()

#분할이 너무 많이 됨 : over segmentation -> diff 높힘
```
