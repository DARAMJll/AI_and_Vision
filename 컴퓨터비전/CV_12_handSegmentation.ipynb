```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
```

### 과제와 연관(심장 주위의 것들을 제거(열림연산?), 정확도에 따라 점수 달라짐)
connected component 로 후처리(원하지 않는 것들을 제거)

```python
src = cv2.imread('./segmentation_rsc/hand.png')
hsv = cv2.cvtColor(src, cv2.COLOR_BGR2HSV) 
lower = np.array([0, 10, 80], dtype = 'uint8') #배열 : hsv값 
upper = np.array([30, 255, 255], dtype = 'uint8')
dst = cv2.inRange(hsv, lower, upper)
```

```python
src = cv2.cvtColor(src, cv2.COLOR_BGR2RGB)
plt.figure(figsize = (10, 10))
image1 = plt.subplot(1, 2, 1)
image1.set_title('Original')
plt.axis('off')
plt.imshow(src)

image2 = plt.subplot(1, 2, 2)
image2.set_title('Detected hand')
plt.axis('off')
plt.imshow(dst, cmap = 'gray')
```

### 손 영역 정확히 따고 싶으면 모폴로지 적용(닫힘연산)

```python
kernel = cv2.getStructuringElement(shape = cv2.MORPH_RECT, ksize = (7, 7))

mor_close = cv2.morphologyEx(dst, cv2.MORPH_CLOSE, kernel, iterations = 5) #5번 // cf : morph_cross 는 십자가 모양
```

```python
plt.figure(figsize = (20, 20))
image1 = plt.subplot(1, 3, 1)
image1.set_title('Original')
plt.axis('off')
plt.imshow(src)

plt.figure(figsize = (20, 20))
image2 = plt.subplot(1, 3, 2)
image2.set_title('Detected hand')
plt.axis('off')
plt.imshow(dst, cmap = 'gray')

plt.figure(figsize = (20, 20))
image3 = plt.subplot(1, 3, 3)
image3.set_title('Final result')
plt.axis('off')
plt.imshow(mor_close, cmap = 'gray')
```
