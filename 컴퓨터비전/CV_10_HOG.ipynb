```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
```

```python
src = cv2.imread('./data/man.png')
gray = cv2.cvtColor(src, cv2.COLOR_BGR2GRAY)
```

```python
#HOG descripter 계산(컬러영상)
hog1 = cv2.HOGDescriptor()
des1 = hog1.compute(src)
print('des1.shape=', des1.shape)
print('des=', des1)
```

```python
winSize     = (64,128)
blockSize   = (16,16)
blockStride = (8,8)
cellSize    = (8,8)
nbins       = 9 #방향 9개
derivAperture = 1  # 미분계산필터크기 (1로 설정하면 3x3)
winSigma = -1      # -1이면 sigma = 4.0
histogramNormType = 0  # L2norm 정규화
L2HysThreshold = 0.2   # 정규화할때 임계값
gammaCorrection = True
nlevels = 64       # detection window 최대값
signedGradient = False  # (0 ~ 360까지 사용하겠다는 의미) / true : 0~180
hog2 = cv2.HOGDescriptor(winSize,blockSize,blockStride,cellSize,nbins,
                         derivAperture,winSigma,
                         histogramNormType,L2HysThreshold,
                         gammaCorrection,nlevels, signedGradient)
des2 = hog2.compute(src)
print('des2.shape=', des2.shape)
print('des2=', des2)
```

```python
#그레이영상
hog3 = cv2.HOGDescriptor()
des3 = hog3.compute(gray)
print('des3.shape = ', des3.shape)
print('des3 = ', des3)
```

```python
plt.title('HOGDescriptor')
plt.plot(des1[::36], color = 'b', linewidth = 4, label = 'des1') #문자열 슬라이싱
plt.plot(des1[::36], color = 'r', linewidth = 4, label = 'des2')
plt.plot(des1[::36], color = 'g', linewidth = 4, label = 'des3')
plt.legend(loc = 'best')
plt.show()
```
