```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
```

```python
src = cv2.imread('./segmentation_rsc/bridge.jpg')
srcrgb = cv2.cvtColor(src, cv2.COLOR_BGR2RGB)
mask = np.zeros((src.shape[0]+2, src.shape[1]+2), np.uint8) #+2를 무조건 해야함(외곽선 패딩)

seed = (180, 80) #height, width
lodiff, updiff = (5, 5, 5), (5, 5, 5)
cv2.floodFill(src, mask, seedPoint = seed, newVal = 255, 
             loDiff = lodiff, upDiff = updiff) #시드와 같은 영역 : 255
cv2.circle(srcrgb, seed, 3, (255,0,0), cv2.FILLED, cv2.LINE_AA) #반지름 3인 원을 색깔로 채워라
```

```python
plt.figure(figsize = (20, 20))
image1 = plt.subplot(1, 2, 1)
image1.set_title('Input image')
plt.axis('off')
plt.imshow(srcrgb)

image1 = plt.subplot(1, 2, 2)
image1.set_title('Result')
plt.axis('off')
plt.imshow(mask, cmap = 'gray')

plt.show()
```

### 마우스 클릭 영역 분할

```python
src2 = cv2.imread('./segmentation_rsc/duck.png')
mask2 = np.zeros((src2.shape[0]+2, src.shape[1]+2), np.uint8)
```

```python
def onMouse(event, x, y, flags, param):
    if event == cv2.EVENT_LBUTTONDOWN:
        seed = (x, y)
        r = np.random.randint(256)
        g = np.random.randint(256)
        b = np.random.randint(256)
        newVal = (b, g, r)
        retval = cv2.floodFill(src2, mask2, seed, newVal = newVal, loDiff = lodiff, upDIff = updiff)
        cv2.imshow('src2', src2)
```

```python
cv2.imshow('src2', src2)
cv2.setMouseCallback('src2', onMouse)
cv2.waitKey()
cv2.destroyAllWindows()
```
