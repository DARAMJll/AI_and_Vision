```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
from skimage import filters #사이킷 이미지(scikit-image) : 기계학습용 라이브러리
```

```python
src = cv2.imread('./segmentation_rsc/mandible.png', cv2.IMREAD_GRAYSCALE) #하악 사진, 이미지 자체가 그레이인데 저장이 칼라로 되어있음
ret, dst = cv2.threshold(src, 128, 255, cv2.THRESH_BINARY + cv2.THRESH_OTSU) #Thresh를 자동 계산
print(f'threshold = {ret}')
```

```python
plt.figure(figsize = (8, 8))
plt.subplot(1, 2, 1) #원본이랑 스레숄드 두 개 뿌림
plt.axis('off')
plt.imshow(src, cmap = 'gray')

plt.subplot(1, 2, 2)
plt.axis('off')
plt.imshow(dst, cmap = 'gray')
```

```python
#참고
thresh = filters.threshold_multiotsu(src, classes = 3) #세개 클래스로 나눠라 -> 쓰레숄드 두 개 필요
ret2, dst2 = cv2.threshold(src, thresh[1], 255, cv2.THRESH_BINARY)
print(f'threshold2 = {ret2}')
```

```python
plt.figure(figsize = (8, 8))
plt.subplot(1, 3, 1) #원본이랑 스레숄드 두 개 뿌림
plt.axis('off')
plt.imshow(src, cmap = 'gray')

plt.subplot(1, 3, 2)
plt.axis('off')
plt.imshow(dst, cmap = 'gray')

plt.subplot(1, 3, 3)
plt.axis('off')
plt.imshow(dst2, cmap = 'gray')
```
