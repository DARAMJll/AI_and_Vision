```python
#데이터셋 만들기
X, y = mglearn.datasets.make_forge()
print(X.shape, y.shape)
print(X, y)

#산점도 그리기
mglearn.discrete_scatter(X[:, 0], X[:, 1], y)
plt.legend(["class 0", "class 1"], loc = 'best')
plt.xlabel("feature 1")
plt.ylabel("feature 2")
print("X.shape:", X.shape)
```
