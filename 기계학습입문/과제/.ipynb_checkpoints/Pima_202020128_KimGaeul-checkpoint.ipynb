```python
import numpy as  np
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
```

```python
input_file = './data/pima-indians-diabetes.data.csv'
data = np.loadtxt(input_file, delimiter = ',')
x, y = data[:,:-1], data[:,-1]
print(data.shape)
print(x[0:5])
print(y[0:5])
x_train, x_test, y_train, y_test = train_test_split(x, y, random_state = 0)
print(x_train.shape)
print(y_train.shape)
print(x_test.shape)
print(y_test.shape)
```

```python
classifier = LogisticRegression(C = 1000, max_iter = 10000).fit(x_train, y_train)
print("훈련 세트 정확도 : {:.2f}%".format(classifier.score(x_train, y_train) * 100))
print("테스트 세트 정확도 : {:.2f}%".format(classifier.score(x_test, y_test) * 100))
#언더피팅 -> x값 높여야 함
```

```python
C = list(np.linspace(0.01, 10, 100))
for c in C:
    logreg = LogisticRegression(C = c, max_iter = 10000).fit(x_train, y_train)
    print("C = ", c)
    print("훈련 세트 정확도 : {:.2f}%".format(logreg.score(x_train, y_train) * 100))
    print("테스트 세트 정확도 : {:.2f}%".format(logreg.score(x_test, y_test) * 100))
    print("")
#모든 값에서 크게 변화가 없다 -> 적합 X
```
